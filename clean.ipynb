```python
import pandas as pd
```

```python
df = pd.read_csv('recent_ac_submissions.csv')
```

```python
df.head()
```

```python
'0.0B'.split()
```

```python
df['runtime_val'] = df['runtime'].apply(lambda x: int(x.split()[0]))
df['memory_val'] = df['memory'].apply(lambda x: float(x.split()[0]))
df.sort_values(by=['username', 'titleSlug', 'runtime_val', 'memory_val'], ascending=[True, True , True, True], inplace=True)
```

```python
df.drop_duplicates(subset=['username', 'titleSlug'], keep='first')[['username', 'titleSlug', 'lang', 'runtime', 'memory', 'timestamp']]

df['overall_score'] = df['runtime_val'] + df['memory_val']
```

```python
df_overall_rank = df.groupby('username')['overall_score'].sum().reset_index().sort_values(by='overall_score')
```
